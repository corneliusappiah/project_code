```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
import scipy 
import ast
```

```python
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df["job_posted_date"]=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_us=df[(df['job_title_short']=="Data Analyst")& (df['job_country']=='United States')].copy()
df_us.dropna(subset=['salary_year_avg'],inplace=True)
```

```python
df_us
```

```python
df_us=df_us.explode('job_skills')
df_us[['salary_year_avg','job_skills']]
```

```python
df_da_group=df_us.groupby('job_skills')['salary_year_avg'].agg(['count','median'])
```

```python
df_da_group
```

```python
df_top_pay=df_da_group.sort_values(by='median',ascending=False).head(10)
df_top_pay
```

```python
df_top_skill=df_da_group.sort_values(by='count',ascending=False).head(10).sort_values(by='median',ascending=False)
df_top_skill
```

```python
fig,ax=plt.subplots(2,1)
#[::-1] is to invert the axis, normally ax[0].invert_yaxis()is used
df_top_pay[::-1].plot(kind='barh',y='median',ax=ax[0],legend=False)
ax[0].set_title('Top 10 Heightest Paid Skills For Data Analyst')
ax[0].set_xlabel('')
ax[0].set_ylabel('')
plt.tight_layout()

df_top_skill.plot(kind='barh',y='median',ax=ax[1],legend=False)
plt.title('Top 10 Most in demand Data Analyst Skills')
plt.xlabel('')
plt.ylabel('median Salary (USD)')
plt.tight_layout()
ax[1].set_xlim(ax[0].get_xlim())
```
