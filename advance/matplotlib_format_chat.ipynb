```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datasets import load_dataset
import ast
```

```python
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df["job_posted_date"]=pd.to_datetime(df['job_posted_date'])
#using lambda
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_skill=df.copy()
df_skill=df_skill.explode('job_skills')
skill_count=df_skill.groupby(['job_skills','job_title_short']).size()
df_skill_count= skill_count.reset_index(name='skill_count')
df_skill_count.sort_values(by="skill_count",ascending=False,inplace=True)
df_skill_count
```

```python
job_titles=['Data Analyst','Data Engineer','Data Scientist']
```

```python
fig,ax=plt.subplots(3,1)
for i,job_title in enumerate(job_titles):
 print(i,job_title)
```

```python
fig,ax=plt.subplots(3,1)
for i,job_title in enumerate(job_titles):
 df_plot=df_skill_count[df_skill_count['job_title_short']==job_title].head(5)
 df_plot.plot(kind='barh',x="job_skills",ylabel="skill_values",ax=ax[i],title=job_title)
 ax[i].invert_yaxis()
 ax[i].set_ylabel('')
 ax[i].legend().set_visible(False)
 ax[i].set_xlim(0,120000)
 fig.suptitle('count of top top skills in job postin',fontsize=15)
 fig.tight_layout()
```
