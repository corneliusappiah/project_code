```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
import scipy 
import ast
```

```python
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df["job_posted_date"]=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_us=df[(df['job_title_short']=='Data Analyst')&(df['job_country']=='United States')]
```

```python
df_us
```

```python
df_us['job_location'].value_counts().head(10)
```

```python
#whant it in data frame 
df_plot=df_us['job_location'].value_counts().head(10).to_frame()
df_plot
```

```python
import seaborn as sns
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,y='job_location',x='count',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Count of job location for Data Analyst in US')
plt.xlabel('Number Of Jobs')
plt.ylabel('')
plt.show()
```

```python

dic_column={
    'job_work_from_home':'Job From Home',
    'job_no_degree_mention':'Job Degree Reg',
    'job_health_insurance':'Health insurance Offered'
}
fig,ax=plt.subplots(1,3)
fig.set_size_inches((12,5))
for i,(column,title) in enumerate(dic_column.items()):
 ax[i].pie(df_us[column].value_counts(), startangle=90,autopct="%1.2f%%",labels=['False','True'])
 ax[i].set_title(title)
plt.show()
```

```python
df_plot=df_us['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,y='company_name',x='count',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Counts of Companies for Data Analyst in US')
plt.xlabel('Number Of Jobs')
plt.ylabel('')
plt.show()
```
