```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datasets import load_dataset
```

```python
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df["job_posted_date"]=pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month']=df["job_posted_date"].dt.strftime('%b')
df
```

```python
month=df['job_posted_month'].unique()
month
```

```python
dict_month={month:df[df['job_posted_month']==month] for month in month}
dict_month['Jan']
```

```python
df_q=pd.concat([dict_month['Jan'],dict_month['Feb'],dict_month['Mar']], ignore_index=True)
```

```python
df_q["job_posted_month"].value_counts().plot(kind="bar")
```
